# 2.5) Boltzmann's Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Initialisation

We have 100 energy equidistant energy levels, each with a  energy from 0 to 99 (in units of kT). First, each of the 10^4 particles a random energy level from 0 to 49 is assigned.

```python
rng = np.random.default_rng()
e_particles = rng.integers(low=0, high=49, size=10**4)
len(e_particles)
```

```python
plt.hist(e_particles, bins=max(e_particles))
plt.xlabel("Energy level")
plt.ylabel("Occupation")
plt.show()
```

## Collisions

Each copmuting step is a collision between particles as descrbed on the problem sheet.

```python
r = 5 #max my computer can do in reasonable time
for i in range(1, 10**r):

    # Choose two random particles with their respective energy lavels

    index_m1, index_m2 = rng.choice(range(0, len(e_particles)), 2, replace=False)
    m1, m2 = [e_particles[index_m1], e_particles[index_m2]]

    #Check if particle energies are min or max

    if m1 == 0 or m2 == 99:
        continue

    else:
        # Increment/Decrement energies

        m1 = m1 - 1
        m2 = m2 + 1
    
    e_particles[index_m1] = m1
    e_particles[index_m2] = m2
    i += 1
```

```python
plt.hist(e_particles, bins = max(e_particles))
plt.xlabel("Energy level")
plt.ylabel("Occupation")
plt.show()
```
